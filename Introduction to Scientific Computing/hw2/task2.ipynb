```python
import numpy as np
import numpy
import seaborn as sns
import math
import matplotlib.pyplot as plt
import seaborn

%matplotlib inline
```

```python
def first_order_derivative(function, point, step=1e-3):
    if step <= 0:
        raise ValueError('step should be positive')
    
    return (function(point + step) - function(point)) / step 

def first_order_derivative_centered(function, point, step=1e-3):
    if step <= 0:
        raise ValueError('step should be positive')
    
    return (function(point + step) - function(point - step)) / 2 / step 

def analytic_derivative(tan):
    return 1./(math.cos(tan))**2
```

```python
def plot_error(approx_function, analytic_function):
    point = 1
    step_sizes = [10**(-k) for k in range(0, 17)]
    errors = [abs(approx_function(math.tan, point, step) - analytic_function(step)) for step in step_sizes]
    plt.figure(figsize=(10, 10))
    plt.xscale('log')
    plt.yscale('log')
    plt.plot(step_sizes, errors)
    return errors, min(errors), max(errors)
```

```python
e1, min1, max1 = plot_error(first_order_derivative, analytic_derivative)
```

```python
e2, min2, max2 = plot_error(first_order_derivative_centered, analytic_derivative)
```

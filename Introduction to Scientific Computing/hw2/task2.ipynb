```python
import numpy as np
import numpy
import seaborn as sns
import math
import matplotlib.pyplot as plt
import seaborn

%matplotlib inline
```

```python
def first_order_derivative(function, point, step=1e-3):
    if step <= 0:
        raise ValueError('step should be positive')
    
    return (function(point + step) - function(point)) / step 

def first_order_derivative_centered(function, point, step=1e-3):
    if step <= 0:
        raise ValueError('step should be positive')
    
    return (function(point + step) - function(point - step)) / 2 / step 

def analytic_derivative(tan):
    return 1./(math.cos(tan))**2
```

```python
def plot_error(approx_function, analytic_function):
    point = 1
    step_sizes = [10**(-k) for k in range(0, 17)]
    errors = [abs(approx_function(math.tan, point, step) - analytic_function(point)) for step in step_sizes]
    plt.figure(figsize=(10, 5))
    plt.xscale('log')
    plt.yscale('log')
    plt.plot(step_sizes, errors)
    return errors, min(errors), max(errors)
```

```python
def prot_errors():
    plt.figure(figsize=(10, 5))
    point = 1
    step_sizes = [10**(-k) for k in range(0, 17)]
    errors1 = [abs(first_order_derivative(math.tan, point, step) - analytic_derivative(point)) for step in step_sizes]
    errors2 = [abs(first_order_derivative_centered(math.tan, point, step) - analytic_derivative(point)) for step in step_sizes]
    
    
    plt.xscale('log')
    plt.yscale('log')
    plt.plot(step_sizes, errors1, label='right')
    plt.plot(step_sizes, errors2, label='centered')
    plt.legend()
    plt.grid(True)
    plt.savefig('./errorplot')
```

```python
prot_errors()
```
